```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import os
import sys
from multiprocessing import Pool

import warnings
warnings.filterwarnings("ignore")

tqdm.pandas()
```

```python
import re
import multiprocessing as mp

def get_mdna(s: str) -> str:
    if pd.isna(s):
        return None
        
    matches = re.findall(r'(?i)MANAGEMENT\'?S? DISCUSSION(?i).*(?i)ITEM[ ]{0,2}\d(?i)', s, re.DOTALL)
    if len(matches) == 1:
        s = matches[0]
    # take the last set of discusion text
    elif len(matches) > 1:
        s = matches[-1]
    return s
```

```python
text_reader = pd.read_csv('all_text.csv', 
                chunksize=1000,
                index_col= 'Unnamed: 0'
            )
text_reader
# for chunk in text_reader:
#     print(chunk.columns)
```

```python
os.path.exists('all_text_mdna.csv')
```

```python
def process_text(df: pd.DataFrame) -> pd.DataFrame:
    # print(mp.current_process().name, 'starting')
    df['text'] = df['text'].apply(get_mdna)
    df.dropna(subset=['text'], inplace=True)
    return df

if os.path.exists('all_text_mdna.csv'):
    os.delete('all_text_mdna.csv')
    
    
with open('all_text_mdna.csv', 'w') as f:
    with tqdm() as pbar:
        with Pool(4) as p:
            for chunk in p.imap(process_text, text_reader):
                pbar.update()
                if os.path.exists('all_text_mdna.csv'):
                    chunk.to_csv('all_text_mdna.csv', mode='a', header=False, index=False)
                else:
                    chunk.to_csv('all_text_mdna.csv', index=False)
```
